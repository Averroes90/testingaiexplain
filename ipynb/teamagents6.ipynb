```python
from utils.credits_tracker import log_usage
from utils.logger_utils import log_anything

team_agent_response = doc_team2.run(query=query)
save_object_to_pickle(team_agent_response.data["output"], f"pkl/html_resume.pkl")
log_anything(team_agent_response, label="team agent response")
log_usage(credits_used=team_agent_response.used_credits)
```
