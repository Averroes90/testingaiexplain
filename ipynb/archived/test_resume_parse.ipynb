```python
from chunking.chunk_merge import nlp

text = "• Completely restructured Clear to Build management process leading to strategic SKU-level visibility Philip Morris International 2013-2017 Operations Planning Analyst (2015-2017) Izmir, Turkey"


def inspect_stanza_features(text: str, nlp) -> None:
    """
    Prints out constituency tree, POS tags, lemmas, dependencies, and entities
    for a given text input using a Stanza pipeline.
    """
    print(f"\n📝 Input Text:\n{text}\n{'='*60}")
    doc = nlp(text)

    for i, sentence in enumerate(doc.sentences):
        print(f"\n🔹 Sentence {i+1}")

        # Constituency tree
        if sentence.constituency:
            print("\n🌲 Constituency Tree:")
            print(sentence.constituency)

        # Token-level features
        print("\n🧠 Tokens + POS + Lemmas:")
        for word in sentence.words:
            print(
                f"  {word.text:<15} POS: {word.upos:<5}  Lemma: {word.lemma:<15} Head: {word.head:<2} Relation: {word.deprel}"
            )

        # Named entities (if present)
        if hasattr(doc, "entities") and doc.entities:
            print("\n🏷️ Named Entities:")
            for ent in doc.entities:
                print(f"  {ent.text:<30} Type: {ent.type:<10}")

    print("=" * 60 + "\n")


inspect_stanza_features(text, nlp)
```
