```python
import pandas as pd
import chunking.chunk_free_form as chunk_free_form

data_url = "my_corpus.csv"
data_df = pd.read_csv(data_url)
data_df
```

```python
first_cell = data_df.loc[0, "content"]
first_cell
```

```python
sents = chunk_free_form.split_into_sentences(first_cell)
sents
```

```python
embeddings = chunk_free_form.embed_sentences(sents)
embeddings
```

```python
similarity_threshold = 0.6
max_tokens = 300
merged_sents = chunk_free_form.merge_sentences_leiden(
    sents,
    embeddings,
    max_token_limit=max_tokens,
    similarity_threshold=similarity_threshold,
)
merged_sents
```

```python
second_cell = data_df.loc[7, "content"]
sents2 = chunk_free_form.split_into_sentences(second_cell)
sents2
embeddings2 = chunk_free_form.embed_sentences(sents2)
merged_sents2 = chunk_free_form.merge_sentences_leiden(
    sents2,
    embeddings2,
    max_token_limit=max_tokens,
    similarity_threshold=similarity_threshold,
)
merged_sents2
```
