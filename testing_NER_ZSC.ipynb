```python
import pandas as pd

data_url = "my_corpus.csv"
data_df = pd.read_csv(data_url)
data_df
```

```python
first_cell = data_df.loc[14, "content"]
first_cell
```

```python
import chunking.chunk_sections as chunk_sections

lines = chunk_sections.parse_into_lines(first_cell)
for i, line in enumerate(lines):
    print(f"{i}: {line}")
```

```python
from nlp.ner import load_ner_model, run_ner_on_chunks

ner_pipeline = load_ner_model()
ner_rusults = run_ner_on_chunks(lines, ner_pipeline)
ner_rusults
```

```python
from nlp.zsc import load_zsc_model, classify_chunks_zsc

zsc_model = load_zsc_model()
zsc_results = classify_chunks_zsc(lines, zsc_model)
```

```python
zsc_results
```
