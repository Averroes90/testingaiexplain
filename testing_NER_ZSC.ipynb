```python
import pandas as pd

data_url = "csv/my_corpus.csv"
data_df = pd.read_csv(data_url)
data_df
```

```python
first_cell = data_df.loc[15, "content"]
first_cell
```

```python
import chunking.chunk_sections as chunk_sections

lines = chunk_sections.parse_into_lines(first_cell)
for i, line in enumerate(lines):
    print(f"{i}: {line}")
```

```python
from nlp.ner import load_ner_model, run_ner_on_chunks

ner_pipeline = load_ner_model()
ner_rusults = run_ner_on_chunks(lines, ner_pipeline)
ner_rusults
```

```python
from nlp.zsc import load_zsc_model, classify_chunks_zsc

zsc_model = load_zsc_model()
zsc_results = classify_chunks_zsc(lines, zsc_model)
```

```python
zsc_results
```

```python
from nlp.rules import annotate_lines

results = annotate_lines(lines)
for res in results:
    print(res)
```

```python
from nlp.rules import compute_perplexity

# comb1 = f"{lines[2]}{lines[3]}{lines[4]}"
comb1 = f"{lines[0]}"
perpcomb = compute_perplexity(comb1)
print(comb1)
perpcomb
```

```python
from nlp.rules import nlp

comb1 = f"{lines[1]} {lines[3]}{lines[4]}"
doc = nlp(comb1)
# print(f"doc {doc}")
# print(f"doc.sent {doc.sentences}")
# Stanza can split your line into multiple sentences. We'll gather the parse for each.
for sent in doc.sentences:  # Convert the internal parse tree to a string
    tree_str = str(sent.constituency)
    # print(f"sent {sent}")
    print(f"sent.cons {sent.constituency}")
    print(f"str(sent.cons) {tree_str}")
```
