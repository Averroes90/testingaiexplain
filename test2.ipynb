```python
import pandas as pd

data_url = "my_corpus.csv"
data_df = pd.read_csv(data_url)
data = data_df.to_dict(orient="records")
```

```python
from aixplain.factories import IndexFactory

# Create an index
index_name = "Test Job Docs"
index_description = "Index for sample job docs"
index = IndexFactory.create(index_name, index_description)
```

```python
from aixplain.modules.model.record import Record

# Prepare the records
records = [
    Record(
        value=item["content"],
        value_type="text",
        id=item["id"],
        uri="",
        attributes={"category": item["doc_type"]},
    )
    for item in data
]

# Upsert records to the index
index.upsert(records)
```
