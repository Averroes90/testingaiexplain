```python
import utils

table1b = utils.load_object_from_pickle("pkl/table1b.pkl")
```

```python
from chunking.combine_features import coarse_segmentation

segs = coarse_segmentation(
    table1b,
    window_size=3,
    threshold=0.34,
    smooth_features=True,
    feature_inclusion={
        "zsc": True,
        "ner": False,
        "formatting": False,
        "embedding": False,
    },
)
print(f"number of segments: {len(segs)+1}")
```

```python
def segment_lines(lines, boundaries):
    segments = []
    start = 0
    for boundary in boundaries:
        segments.append(lines[start:boundary])
        start = boundary
    segments.append(lines[start:])  # Last segment
    return segments


segments = segment_lines(table1b, segs)
print("Number of segments:", len(segments))
for i, segment in enumerate(segments):
    print(f"Segment {i}:")
    for line in segment:
        print(f'  {line["line"]}')
```
