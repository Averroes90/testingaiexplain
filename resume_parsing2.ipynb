```python
import chunking.chunk_sections as chunk_sections
import pandas as pd

data_url = "csv/my_corpus.csv"
data_df = pd.read_csv(data_url)
data_df
```

```python
first_cell = data_df.loc[18, "content"]
first_cell
```

```python
lines = chunk_sections.parse_into_lines(first_cell)
for i, line in enumerate(lines):
    print(f"{i}: {line}")
```

```python
table1a = chunk_sections.generate_line_features(lines)
table1a
```

```python
from chunking.chunk_merge import merge_line_chunks

merged_lines1 = merge_line_chunks(table1a)
```

```python
for i, line in enumerate(merged_lines1):
    print(f"{i}: {line}")
```

```python
table2a = chunk_sections.generate_line_features(lines)
first_cell = data_df.loc[16, "content"]
lines = chunk_sections.parse_into_lines(first_cell)
table2a = chunk_sections.generate_line_features(lines)
merged_lines2 = merge_line_chunks(table2a)
for i, line in enumerate(merged_lines2):
    print(f"{i}: {line}")
```

```python
from itertools import zip_longest

for index, (line1, line2) in enumerate(
    zip_longest(merged_lines1, merged_lines2, fillvalue="")
):
    print(f"{index}: {line1}")
    print(f"{index}: {line2}")
```

```python
table1b = chunk_sections.generate_line_features(merged_lines1)
table1b
```

```python
table2b = chunk_sections.generate_line_features(merged_lines2)
table2b
```
